```python
# Importando as bibliotecas 

import numpy as np
import pandas as pd
import datetime as dt
import yfinance as yf
import seaborn as sns
import plotly as pl 

%matplotlib inline
```

```python
# Utilizando o módulo yfinance para encontrar o ticker que corresponde as ações do twitter

ticker = 'TWTR34.SA'

df = yf.download(ticker, start='2013-11-07')

df
```

```python
# Preço das ações ajustado

df["Adj Close"]
```

```python
# armazendo em uma variável

precos = df["Adj Close"]
```

```python
# plotanto o preço de fechamento ajustado

plt.figure(figsize=(14, 6))
plt.title(f"{ticker} - preço de fechamento ajustado")

# plot com o pandas
precos.plot()

plt.show()
```

```python
# Preço de fechamento
df["Close"].plot(figsize=(20, 5), label="TWTR34.SA")

# Média móvel 21 dias
df["Close"].rolling(21).mean().plot(label="mm-21")

# Média móvel 200 dias
df["Close"].rolling(200).mean().plot(label="mm-200")

# Legendas
plt.legend()
plt.show()
```
